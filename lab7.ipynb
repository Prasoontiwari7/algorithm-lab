```python
def fractional_knapsack(values, weights, capacity):

    items = []
    for i in range(len(values)):
        if weights[i] > 0:
            items.append({'ratio': values[i] / weights[i], 'v': values[i], 'w': weights[i], 'id': i})


    items.sort(key=lambda x: x['ratio'], reverse=True)

    total_value = 0.0


    for item in items:
        if capacity == 0:
            break
        if item['w'] <= capacity:
            total_value += item['v']
            capacity -= item['w']
        else:

            total_value += item['ratio'] * capacity
            capacity = 0

    return total_value


values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50

max_value = fractional_knapsack(values, weights, capacity)
print(f"Max Value: {max_value:.2f}")




```

```python
import heapq
from collections import Counter

class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
    def __lt__(self, other):
        return self.freq < other.freq

def build_tree(text):
    freq = Counter(text)
    pq = [Node(c, f) for c, f in freq.items()]
    heapq.heapify(pq)
    while len(pq) > 1:
        l, r = heapq.heappop(pq), heapq.heappop(pq)
        m = Node(None, l.freq + r.freq)
        m.left, m.right = l, r
        heapq.heappush(pq, m)
    return pq[0]

def get_codes(node, code="", codes={}):
    if node:
        if node.char is not None: codes[node.char] = code
        get_codes(node.left, code + "0", codes)
        get_codes(node.right, code + "1", codes)
    return codes

if __name__ == "__main__":
    text = "huffman coding"
    root = build_tree(text)
    codes = get_codes(root)
    encoded = "".join(codes[c] for c in text)
    print(encoded)
```
