```python
def is_even(n):
    return n % 2 == 0

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    is_even(n)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity (Even/Odd)")
plt.show()
```

```python
def count_positive(arr):
    count = 0
    for x in arr:
        if x > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(-n, n))
    count_positive(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
def binary_search(arr, key):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

n_values = [2**i for i in range(1, 8)]
time_values = []

for n in n_values:
    data = list(range(n))
    binary_search(data, n-1)
    time_values.append(len(bin(n)))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(log n) Logarithmic Time Complexity")
plt.show()
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n, 0, -1))
    merge_sort(data)
    time_values.append(n * (n.bit_length()))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n log n) Merge Sort")
plt.show()
```

```python
def quadratic_example(arr):
    count = 0
    for i in arr:
        for j in arr:
            count += 1
    return count

n_values = list(range(1, 51))
time_values = []

for n in n_values:
    data = list(range(n))
    quadratic_example(data)
    time_values.append(n*n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n²) Quadratic Time Complexity")
plt.show()
```

```python
def cubic_example(arr):
    count = 0
    for i in arr:
        for j in arr:
            for k in arr:
                count += 1
    return count

n_values = list(range(1, 21))
time_values = []

for n in n_values:
    data = list(range(n))
    cubic_example(data)
    time_values.append(n**3)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n³) Cubic Time Complexity")
plt.show()
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n, 0, -1))
    quick_sort(data)
    time_values.append(n * (n.bit_length()))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("Quick Sort – O(n log n)")
plt.show()
```
