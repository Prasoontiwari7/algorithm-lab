```python
import time
start=time.time()
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
    end=time.time()
  print(end-start)
  return result
```

```python
# Sample function call: func2(2, 4)
print(func2(2,4))
#to print a raise to power b
```

```python
# Exercise 4
import time
start=time.time()
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    end=time.time()
    print(end-start)
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
print(func4([1, 2, 3, 4], [2, 1, 4, 3]))
# returns the number of common elements in both the list
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1

    return None
```

```python
# Sample function call: func5(16)
import time
start=time.time()
print(func5(25))
end=time.time()
print(end-start)
# gives square root of the number
```

```python
# Exercise 6
import time
start=time.time()
def func8(A):
    func8_helper(A, 0, len(A) -1)
    end=time.time()
    print(end-start)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A

```

```python
#Sample function call: func8([3,1,2,7,8,6])
print(func8([3,1,2]))
```

```python
# Exercise 7
import time
start=time.time()
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    end=time.time()
    print(end-start)
    return custom_arr


```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
print(func12([1, 2, 3, 4, 5], 2))
#shifting from the end as the input is given
```

```python
def sum(arr):
  sum = 0
  for i in range(len(arr)):
    sum += arr[i]
  return sum
```

```python
print(sum([12, 3, 4, 15,45,78,450]))
```

```python
def max(arr):
  max = arr[0]
  for i in range(len(arr)):
    if arr[i] > max:
      max = arr[i]
  return max

```

```python
print(max([12, 3, 4, 15,45,78,450]))
```
