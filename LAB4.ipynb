```python
import time
import random
import matplotlib.pyplot as plt

def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

input_sizes = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
time_taken = []

for n in input_sizes:
    data = [random.randint(0, n) for _ in range(n)]
    search_element = random.randint(0, n)

    start_time = time.time()
    linear_search(data, search_element)
    end_time = time.time()
    time_taken.append((end_time - start_time) * 1000)

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, time_taken, marker='o')
plt.title('Linear Search Time Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Time Taken (milliseconds)')
plt.grid(True)
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# Recursive Binary Search function
def recursive_binary_search(arr, low, high, x):
    if high >= low:
        mid = (high + low) // 2

        # If element is present at the middle itself
        if arr[mid] == x:
            return mid

        # If element is smaller than mid, then it can only be present in left subarray
        elif arr[mid] > x:
            return recursive_binary_search(arr, low, mid - 1, x)

        # Else the element can only be present in right subarray
        else:
            return recursive_binary_search(arr, mid + 1, high, x)
    else:
        # Element is not present in the array
        return -1

# Input sizes to test
input_sizes = [1000, 5000, 10000, 50000, 100000, 200000, 500000, 1000000]
time_taken = []

for n in input_sizes:
    # Generate a sorted list of 'n' random numbers
    data_list = sorted([random.randint(0, n * 2) for _ in range(n)])
    # Choose an element to search for (might or might not be in the list)
    search_element = random.randint(0, n * 2)

    start_time = time.time()
    recursive_binary_search(data_list, 0, len(data_list) - 1, search_element)
    end_time = time.time()
    time_taken.append((end_time - start_time) * 1000) # Convert to milliseconds

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, time_taken, marker='o')
plt.title('Recursive Binary Search Time Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Time Taken (milliseconds)')
plt.grid(True)
plt.xscale('log') # Use log scale for x-axis if input sizes vary greatly
plt.yscale('log') # Use log scale for y-axis if time taken varies greatly
plt.show()
```

```python
def search(pat, txt):
    n = len(txt)
    m = len(pat)
    occurrences = []

    # 2. Loop through each character in the text
    # We stop at n - m + 1 because the pattern can't fit in the remaining text after that
    for i in range(n - m + 1):

        # Check if the substring starting at i matches the pattern
        if txt[i : i + m] == pat:
            # 3. If match, append the starting index
            print(f"Pattern found at index {i}")
            occurrences.append(i)

    # 5. Return the list of indices
    return occurrences

# Example Usage:
text = "AABAACAADAABAABA"
pattern = "AABA"
search(pattern, text)
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        # Compare and shift elements
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Benchmarking
sizes_sort = [100, 500, 1000, 2000, 3000]
sort_times = []

for n in sizes_sort:
    # Use reverse sorted list for worst-case scenario
    test_arr = list(range(n, 0, -1))
    start = time.perf_counter()
    insertion_sort(test_arr)
    sort_times.append(time.perf_counter() - start)

plt.plot(sizes_sort, sort_times, marker='s', color='red')
plt.title("Insertion Sort Time Complexity (O(n))")
plt.xlabel("Number of Elements (n)")
plt.ylabel("Time (seconds)")
plt.show()
```
