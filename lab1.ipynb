```python
def gcd_simple(m, n):
    m = abs(m)
    n = abs(n)
    gcd = 1
    for i in range(1, min(m, n) + 1):
        if m % i == 0 and n % i == 0:
            gcd = i
    return gcd
```

```python
start_time = timeit.default_timer()
g = gcd_simple(m, n)
elapsed = (timeit.default_timer() - start_time) * 1000
print("GCD:", g)
print("Elapsed time:", elapsed, "ms")
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```
